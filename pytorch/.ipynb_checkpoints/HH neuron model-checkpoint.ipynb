```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#implementation of HH model for single neuron
```

```python
#first, define voltage dependent gate variable
def alpha_m(V):
    return (2.5-0.1*V) / (np.exp(2.5-0.1*V) -1)

def beta_m(V):
    return 4*np.exp(-V/18)

def alpha_h(V):
    return 0.07*np.exp(-V/20)

def beta_h(V):
    return 1/(np.exp(3.0-0.1*V)+1)

def alpha_n(V):
    return (0.1-0.01*V) / (np.exp(1-0.1*V) -1)

def beta_n(V):
    return 0.125*np.exp(-V/80)
```

```python
def pulse_generator(i, p, t, N, dt = 0.001):
    impulse = np.zeros(N)
    no_cycle = (N*dt) // t
    sample_per_cycle = int(t/dt)
    for j in range(int(no_cycle)):
        start_ind = int(sample_per_cycle * (j+1) - p*1000)
        impulse[start_ind: (j+1)*sample_per_cycle] = 1
    return impulse*i
```

```python
def update(i_ext,m,v,h,n,dt = 0.001):
    vn = v + dt*(-g_na*m**3 * h * (v-e_na) - g_k*n**4 * (v-e_k) - g_l * (v-e_l) + i_ext)
    mn = m + dt*(alpha_m(v) * (1-m) - beta_m(v) * m)
    hn = h + dt*(alpha_h(v) * (1-h) - beta_h(v) * h)
    nn = n + dt*(alpha_n(v) * (1-n) - beta_n(v) * n)
    return vn, mn, hn, nn
```

```python
g_na = 120
g_k = 36
g_l = 0.3
e_na = 115
e_k = -12
e_l = 10.6
```

```python
v_t = np.linspace(-50,120,1000)
plt.plot(v_t, alpha_m(v_t)/(alpha_m(v_t) + beta_m(v_t)))
plt.plot(v_t, alpha_n(v_t)/(alpha_n(v_t) + beta_n(v_t)))
plt.plot(v_t, alpha_h(v_t)/(alpha_h(v_t) + beta_h(v_t)))
plt.legend(['m','n','h'])
print(alpha_m(0)/(alpha_m(0) + beta_m(0)))
print(alpha_n(0)/(alpha_n(0) + beta_n(0)))
print(alpha_h(0)/(alpha_h(0) + beta_h(0)))
```

```python
#simulation 1: constant input
v, m, n, h = 0, 0.053, 0.318, 0.596
i_ext = 8
N = 200000
v_array = np.zeros(N)
m_array = np.zeros(N)
h_array = np.zeros(N)
n_array = np.zeros(N)
for i in range(N):
    v,m,h,n = update(i_ext,m,v,h,n,dt = 0.001)
    v_array[i] = v
    m_array[i] = m
    h_array[i] = h
    n_array[i] = n
```

```python
plt.plot(10**-3*np.arange(N),v_array)
```

```python
plt.plot(10**-3*np.arange(N),m_array)
plt.plot(10**-3*np.arange(N),h_array)
```

```python
def fre_test(i_range):
    firing_rate = np.zeros(i_range.shape)
    j = 0
    for i_test in i_range:
        v, m, n, h = 0, 0.053, 0.318, 0.596
        i_ext = 50
        N = 200000
        v_array = np.zeros(N)
        for i in range(N):
            v,m,h,n = update(i_test,m,v,h,n,dt = 0.001)
            v_array[i] = v
        firing_rate[j] = np.argmax(np.abs(np.fft.fft(v_array-np.mean(v_array)))[0:300])
        j = j+1
    return firing_rate
```

```python
i_range = np.linspace(5,80,40)
plt.plot(i_range,fre_test(i_range))
```

```python
#simulation 2: constant input
v, m, n, h = 0, 0.053, 0.318, 0.596
```

```python
i = -5
p = 8
t = 40
N = 200000

i_pulse = pulse_generator(i, p, t, N, dt = 0.001)

v_array = np.zeros(N)
m_array = np.zeros(N)
h_array = np.zeros(N)
n_array = np.zeros(N)

for i in range(N):
    v,m,h,n = update(i_pulse[i],m,v,h,n,dt = 0.001)
    v_array[i] = v
    m_array[i] = m
    h_array[i] = h
    n_array[i] = n
```

```python
plt.plot(10**-3*np.arange(N),v_array)
```
