```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
sigma = np.zeros((n,n))
for i in range(n*2):
    gi = np.random.randn(n,1)
    sigma = sigma + gi@gi.T
sigma = sigma/n
```

```python
s = np.random.randn(n,n)
for i in range(n):
    s[i,i] = 0
    for j in range(i):
        s[i,j] = -s[j,i]
```

```python
plt.plot(np.linalg.eig(sigma)[0])
```

```python
def update(r, W, delta_t,tau_m = 0.02):
    dr = delta_t/tau_m*(-r+W@r)+delta_t**0.5*(2/tau_m)**0.5*np.random.randn(r.shape[0],r.shape[1])
    return r+dr
```

```python
#Langevin: W = \Sigma^-1
W = np.eye(n)-np.linalg.inv(sigma)
r = np.random.randn(n,1)
delta_t = 0.00001
l_r = []

for i in range(5000000):
    r = update(r, W, delta_t)
    r_a = r.copy()
    l_r.append(r_a.reshape(-1))

l_r = np.array(l_r)
```

```python
plt.imshow(l_r.T[:,:50000],aspect='auto')
```

```python
W1 = np.eye(n) + (-np.eye(n)+s)@np.linalg.inv(sigma)
r = np.random.randn(n,1)
delta_t = 0.00001
l_r1 = []

for i in range(5000000):
    r = update(r, W1, delta_t)
    r_a = r.copy()
    l_r1.append(r_a.reshape(-1))

l_r1 = np.array(l_r1)
```

```python
plt.plot(np.linalg.eig(W1)[0]-1)
plt.plot(np.linalg.eig(W)[0]-1)
plt.legend(['non sym','lengevin'])
```

```python
plt.imshow(l_r1.T[:,:50000],aspect='auto')
```

```python
print(np.mean(l_r,axis = 0))
```

```python
def cal_cov(l_r,n):
    cov = np.zeros((n,n))
    mu = np.mean(l_r,axis = 0)
    for i in range(l_r.shape[0]):
        cov = cov + (l_r[0]-mu).reshape(n,1)@(l_r[0]-mu).reshape(1,n)
    return cov/l_r.shape[0]
```

```python
plt.imshow(np.cov(l_r1.T)-sigma)
plt.colorbar()
```

```python
import scipy.linalg
tl = np.linspace(0,10,100)

def cal_K(t,W,sigma,n):
    return scipy.linalg.expm(t*(W-np.eye(n)))@sigma
```

```python
Kl = []
for i in tl:
    Kl.append(scipy.linalg.norm(cal_K(i,W,sigma,n),ord = 'fro'))
Kl = np.array(Kl)

Kl1 = []
for i in tl:
    Kl1.append(scipy.linalg.norm(cal_K(i,W1,sigma,n),ord = 'fro'))
Kl1 = np.array(Kl1)
```

```python
plt.plot(tl,Kl)
plt.plot(tl,Kl1)
```
