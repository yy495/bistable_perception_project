```python
import numpy as np
from matplotlib import pyplot as plt
import torch
```

```python
#Fitzhugh-Nagumo system
```

```python
def update(v,r,i,dt):
    return 10*dt*(-v**3/3+v-r+i)+v, dt*(v-0.8*r+1.2)+r
```

```python
x = np.linspace(-3,3,100)
plt.plot(x,-x**3/3+x)
plt.plot(x,1.25*x+1.5)
```

```python
v_array = np.zeros(4000)
r_array = np.zeros(4000)
v = -2
r = 4
i = 2
dt = 0.01
for p in range(4000):
    v,r = update(v,r,i,dt)
    v_array[p] = v
    r_array[p] = r
```

```python
print(v_array[-1],r_array[-1])
```

```python
plt.scatter(v_array,r_array)
```
