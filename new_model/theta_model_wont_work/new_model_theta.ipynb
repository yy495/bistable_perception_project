```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_heat_map_x(kappa,kappa_0,prior):
    X_L = np.arange(-np.pi,np.pi,0.01)
    X_R = np.arange(-np.pi,np.pi,0.01)
    X_L, X_R = np.meshgrid(X_L,X_R)
    P_00_raw = prior[0]*np.ones(X_L.shape)
    P_01_raw = prior[1]*np.i0(np.abs(kappa*np.exp(1j*X_R)+kappa_0))/np.i0(kappa)/np.i0(kappa_0)
    P_10_raw = prior[2]*np.i0(np.abs(kappa*np.exp(1j*X_L)+kappa_0))/np.i0(kappa)/np.i0(kappa_0)
    P_11_raw = prior[3]*np.i0(np.abs(kappa*np.exp(1j*X_R)+kappa*np.exp(1j*X_L)+kappa_0))/np.i0(kappa)**2/np.i0(kappa_0)
    P00 = P_00_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P01 = P_01_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P10 = P_10_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P11 = P_11_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    return P00,P01,P10,P11
```

```python
P00,P01,P10,P11 = plot_heat_map_x(5,0.4,[0.04,0.16,0.16,0.64])
plt.rcParams["figure.figsize"] = (12,12)
plt.subplot(221)
plt.imshow(np.flip(P00,0),extent = [-np.pi,np.pi,-np.pi,np.pi])
plt.colorbar()
plt.xlabel('x_l')
plt.ylabel('x_r')
plt.title('p(z_l,z_r= 0,0|...)')
plt.subplot(222)
plt.imshow(np.flip(P01,0),extent = [-np.pi,np.pi,-np.pi,np.pi])
plt.colorbar()
plt.xlabel('x_l')
plt.ylabel('x_r')
plt.title('p(z_l,z_r = 0,1|...)')
plt.subplot(223)
plt.imshow(np.flip(P10,0),extent = [-np.pi,np.pi,-np.pi,np.pi])
plt.colorbar()
plt.xlabel('x_l')
plt.ylabel('x_r')
plt.title('p(z_l,z_r = 1,0|...)')
plt.subplot(224)
plt.imshow(np.flip(P11,0),extent = [-np.pi,np.pi,-np.pi,np.pi])
plt.colorbar()
plt.xlabel('x_l')
plt.ylabel('x_r')
plt.title('p(z_l,z_r = 1,1|...)')
```

```python
def plot_heat_map_theta(x_l,x_r,kappa_0):
    kappa_l = np.arange(0,10,0.01)
    kappa_r = np.arange(0,10,0.01)
    kappa_l,kappa_r = np.meshgrid(kappa_l,kappa_r)
    P_00_raw = np.ones(kappa_l.shape)
    P_01_raw = np.i0(np.abs(kappa_r*np.exp(1j*x_r)+kappa_0))/np.i0(kappa_r)/np.i0(kappa_0)
    P_10_raw = np.i0(np.abs(kappa_l*np.exp(1j*x_l)+kappa_0))/np.i0(kappa_l)/np.i0(kappa_0)
    P_11_raw = np.i0(np.abs(kappa_l*np.exp(1j*x_l)+kappa_r*np.exp(1j*x_r)+kappa_0))/np.i0(kappa_l)/np.i0(kappa_r)/np.i0(kappa_0)
    P00 = P_00_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P01 = P_01_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P10 = P_10_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    P11 = P_11_raw/(P_00_raw+P_01_raw+P_10_raw+P_11_raw)
    return P00,P01,P10,P11
```

```python
P00,P01,P10,P11 = plot_heat_map_theta(np.pi/4,-np.pi/4,0)
```

```python
plt.rcParams["figure.figsize"] = (12,12)
plt.subplot(221)
plt.imshow(np.flip(P00,0),extent = [0,10,0,10])
plt.colorbar()
plt.xlabel('kappa_l')
plt.ylabel('kappa_r')
plt.title('p(z_l,z_r= 0,0|...)')
plt.subplot(222)
plt.imshow(np.flip(P01,0),extent = [0,10,0,10])
plt.colorbar()
plt.xlabel('kappa_l')
plt.ylabel('kappa_r')
plt.title('p(z_l,z_r = 0,1|...)')
plt.subplot(223)
plt.imshow(np.flip(P10,0),extent = [0,10,0,10])
plt.colorbar()
plt.xlabel('kappa_l')
plt.ylabel('kappa_r')
plt.title('p(z_l,z_r = 1,0|...)')
plt.subplot(224)
plt.imshow(np.flip(P11,0),extent = [0,10,0,10])
plt.colorbar()
plt.xlabel('kappa_l')
plt.ylabel('kappa_r')
plt.title('p(z_l,z_r = 1,1|...)')
```
