```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,5,0.01)
Y = np.arange(0,5,0.01)
X,Y = np.meshgrid(X,Y)
Z = np.i0((X**2+Y**2)**0.5)/(np.i0(X)*np.i0(Y))
plt.imshow(Z,extent=[0,5,5,0])
plt.colorbar()
plt.xlabel('alpha_l')
plt.ylabel('alpha_r')
```

```python
print(np.i0(18**0.5)/np.i0(3)**2)
print(np.i0(32**0.5)/np.i0(4)**2)
print(np.i0(50**0.5)/np.i0(5)**2)
```

```python
X = np.arange(-np.pi/4,np.pi/4,0.01)
Y = np.arange(-np.pi/4,np.pi/4,0.01)
X,Y = np.meshgrid(X,Y)
Z = np.i0(np.abs(3*np.exp(1j*X)+3*np.exp(1j*Y)))/np.i0(3)**2
plt.imshow(Z,extent=[-np.pi/4,np.pi/4,np.pi/4,-np.pi/4])
plt.xlabel('x_l')
plt.ylabel('x_r')
plt.colorbar()
```
