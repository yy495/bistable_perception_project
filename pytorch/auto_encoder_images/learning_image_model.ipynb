```python
import numpy as np
from matplotlib import pyplot as plt
import torch
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
A = unpickle('data_batch_2')
```

```python
image_set = list(A.items())[2][1]
```

```python
im2 = np.zeros((32,32,3))
R = image_set[5,:1024].reshape(32,32)
G = image_set[5,1024:2048].reshape(32,32)
B = image_set[5,2048:].reshape(32,32)
for i in range(32):
    for j in range (32):
        im2[i,j] = np.array([R[i,j],G[i,j],B[i,j]])
im2 = im2.astype('int')
plt.imshow(im2)
```

```python
image_set_grayscale = 0.2126*image_set.reshape(10000,3,1024)[:,0,:] + 0.7152*image_set.reshape(10000,3,1024)[:,1,:] + 0.0722*image_set.reshape(10000,3,1024)[:,2,:]
image_set_grayscale = (image_set_grayscale.T).astype('float32')
image_set_grayscale = torch.tensor(image_set_grayscale)/256
```

```python
image_set_grayscale.shape
```

```python
plt.imshow(image_set_grayscale[:,5].reshape(32,32),cmap='Greys')
```

```python
class network1():
    def forward(W1,x):
        x_r = W1@torch.linalg.inv(W1.T@W1)@W1.T@x
        l = torch.sum(torch.var(x_r-x,axis = 1)) + 0.01*torch.sum(torch.sum(torch.abs(W1.T@x))/10000)
        return l
```

```python
W = torch.autograd.Variable(torch.randn(1024,81),requires_grad = True)
```

```python
optimizer = torch.optim.Adam([W], 0.1)

#loss_list = []
for i in range(1000):
    L = network1.forward(W,image_set_grayscale)
    optimizer.zero_grad()
    L.backward()
    optimizer.step()
    L1 = L.detach().numpy().copy()
    loss_list.append(L1)
```

```python
plt.rcParams["figure.figsize"] = (6,6)
plt.plot(np.array(loss_list))
print(loss_list[-1])
```
